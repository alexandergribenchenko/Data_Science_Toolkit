# 01. gropupby uses

```python
import pandas as pd
```

```python
dict_input = { 'col_01': ['a','b','a','b'],
               'col_02': ['gato','perro','gato','mico'],
               'col_03': [1,5,6,7]
             }

df = pd.DataFrame.from_dict(dict_input)
df
```

```python
df.\
groupby(['col_01', 'col_02'],
  as_index=True
).\
agg(min_col_3=('col_03', 'min'),
    count_col_3=('col_03', 'count'),
    unique_col_2=('col_02', lambda x: x.nunique()),
    conca=('col_02', lambda x: ','.join(x))
).\
sort_values(by=['conca','col_01'],
            ascending=[False,True]
)
```

```python
df.\
groupby(['col_01', 'col_02'],
  as_index=False
).\
agg(min_col_3=('col_03', 'min'),
    count_col_3=('col_03', 'count'),
    unique_col_2=('col_02', lambda x: x.nunique()),
    conca=('col_02', lambda x: ','.join(x))
).\
sort_values(by=['conca','col_01'],
            ascending=[False,True]
)
```

# 03. pivot uses

```python
import pandas as pd
```

```python
dict_input = {"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]
             }

df = pd.DataFrame.from_dict(dict_input)
df
```

```python
df.pivot_table(index = ['A','C'], columns=['B'], values =['D','E'], aggfunc ='sum').reset_index()
```

# 04. Pivot con agregado de columnas

```python
import pandas as pd
import numpy as np

# Crear datos de ejemplo
np.random.seed(0)
data = {
    'probability': np.random.choice([0.25, 0.5, 0.75, 0.9], size=100),
    'fleet_size': np.random.randint(1, 5, size=100),
    'month': np.random.randint(1, 13, size=100),
    'amount': np.random.randint(10, 500, size=100)
}

# Crear DataFrame a partir del diccionario
df_salida = pd.DataFrame(data)
df_salida
```

```python
df_min_max = df_salida.pivot_table(index = ['fleet_size','month'], columns=['probability'], values =['amount'], aggfunc =['min','max']).reset_index()
df_min_max
```

```python
df_min_max.columns
```

```python
df_min_max.columns = ['_'.join(filter(None, map(str, col))) for col in df_min_max.columns]
```
