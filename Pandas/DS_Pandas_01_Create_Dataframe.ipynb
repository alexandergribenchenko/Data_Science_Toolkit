# 04. Save Dataframe to: xlsx file (Drive, Github, Databricks-Datalake)

```python
import pandas as pd
```

```python
import xlsxwriter
```

```python
!pip install xlsxwriter
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
a = [1,2,3,4,5]
b = [6,7,8,9,10]

df = pd.DataFrame(list(zip(a,b)), 
                  columns=['col_01','col_02'])
df
```

```python
path_drive = '/content/drive/MyDrive/Colab Notebooks/df_output_excel.xlsx'
path_databricks_datalake = '/dbfs/mnt/'+'adv/Network_Planning/Caso_02/Sample_Input_Data/df_output_excel.xlsx'
```

```python
sheet_name_01 = "Hoja_01"
sheet_name_02 = "Hoja_02"
```

```python
with pd.ExcelWriter(path_drive, engine='xlsxwriter') as writer: 
  # Escritura del libro
  df.to_excel(writer, sheet_name=sheet_name_01, index = False)
  df.to_excel(writer, sheet_name=sheet_name_02, index = False)
  # Formato del libro
  workbook  = writer.book
  worksheet = writer.sheets[sheet_name_01]
  header_format = workbook.add_format({'bold': True, 'fg_color': '#FF0000','font_color': 'white', 'border': 0})
  for col_num, value in enumerate(df.columns.values):
      worksheet.write(0, col_num, value, header_format)
```

# 08. Create Dataframe from: series

```python
import pandas as pd
```

```python
author = ['Jitender', 'Purnima', 'Arpit', 'Jyoti']
article = [210, 211, 114, 178]
  
auth_series = pd.Series(author)
article_series = pd.Series(article)
  
frame = {'Author': auth_series, 'Article': article_series}
  
df = pd.DataFrame(frame)
df
```
